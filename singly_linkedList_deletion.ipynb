```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
           
        
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
        
    def insert_at_begin(self,data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
     
    def insert_at_end(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            itr = self.head
            while itr.next:
                itr = itr.next
            itr.next = new_node    
    
    def insert_at_position(self,index,data):
        if index < 0 :
            print("inde out of range:")
            return
        if index == 0:
            self.insert_at_begin(data)
            return
        new_node = Node(data)
        current = self.head
        prev = None
        count = 0
        while current and count < index:
            prev = current
            current = current.next
            count +=1
        if count <index:
            print("index is out of the range;")
            return
        prev.next = new_node
        new_node.next = current
            
    def delete_first(self):
        if self.head is None:
            print("linked list is empty:")
            return
        self.head = self.head.next
     
    def delete_at_last(self):
        if self.head is None:
            print("list is empty:")
            return
        if self.head.next is None:
            self.head = None
            return
        
        current = self.head
        prev = None
        while current.next:
            prev = current
            current = current.next
        prev.next = None    
                   
    def delete_at_position(self,index):
        if index < 0 :
            print("inde out of range:")
            return
        if index == 0:
            self.delete_first()
            return

        current = self.head
        prev = None
        count = 0
        while current and count < index:
            prev = current
            current = current.next
            count +=1
        if count <index:
            print("index is out of the range;")
            return
        prev.next = current.next

        
        
    def funct_print(self):
        if self.head is None:
            print("Linked List is empty:")
            return
        
        strll = ''
        itr = self.head
        while itr:
            strll += str(itr.data)+"-->"
            itr = itr.next
        print(strll)    
```
